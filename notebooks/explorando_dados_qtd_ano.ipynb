```python
#quero ler o csv Comercio.csv
df_Comercio = pd.read_csv("../data/dados_raws/Comercio.csv", sep=";", header=None)
dict_Comercio_colunas_novos = {1: 'Tipo', 2: 'Produto'}

#criando um dicionario para renomear as colunas
for ind, ano in enumerate(range(1970, 2022)):
  dict_Comercio_colunas_novos[ind + 3] = ano

#renomeando as colunas e removendo a primeira linha
df_Comercio = df_Comercio.rename(dict_Comercio_colunas_novos, axis=1).drop(0,axis=1)
df_Comercio.head()
```

```python
lista_anos = list(range(1970, 2022))
lista_anos[:5]
```

```python
df_Comercio_melt.to_csv("../data/dados_tratados/Comercio.csv", index=False, encoding='utf-8-sig', sep=';')
```
