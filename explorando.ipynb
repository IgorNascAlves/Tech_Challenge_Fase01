```python
url_ExpVinho = 'https://raw.githubusercontent.com/IgorNascAlves/Tech_Challenge_Fase01/main/tech-challenge/content/ExpVinho.csv'
```

```python
import pandas as pd
```

```python
df_ExpVinho = pd.read_csv(url_ExpVinho, sep=';', header=0)
df_ExpVinho
```

Precisamos identificar de alguma maneira a coluna quantidade(KG) e valor(US$)

image.png

```python
colunas_nomes_anos = df_ExpVinho.columns[2:]
colunas_nomes_anos
```

```python
dict_colunas_nomes = {chave: chave[:4] + '_Quantidade (Kg)' if not ind%2 else chave[:4] + '_Valor (US$)' for ind, chave in enumerate(colunas_nomes_anos) }
# dict_colunas_nomes
```

```python
df_ExpVinho.rename(dict_colunas_nomes, axis=1)
```

image.png

```python
df_ExpVinho.rename(dict_colunas_nomes, axis=1).iloc[:,[0,1,-2,-1]]
```

As informações estãos corretas de acordo com o PDF de referência
